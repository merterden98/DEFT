```python
import pandas as pd
import sys
sys.path.append("../")
from cli import Predict
from glob import glob
FILE = "recon_microbe.tsv"
TRAIN_DB = "/hpc/group/singhlab/rawdata/afdbMert/clean70_0/clean70_db"
TRAIN_CSV = "/hpc/group/singhlab/rawdata/afdbMert/clean70_0/clean70_0_res.csv"
output = "/hpc/group/singhlab/rawdata/afdbMert/vih"
model = "../models/"
PEFT = "/hpc/group/singhlab/rawdata/afdbMert/clean70_0_model/"

df = pd.read_csv(FILE, sep="\t")
num = 0
total = 0
for i, row in df.iterrows():
    try:
        id = int(row["ncbiid"])
        f = glob(f"{output}/{id}*aln.m8")
        if len(f) > 0:
            num += 1
            data = f"{output}/{id}/{id}_res.csv"
            # Predict
            if len(glob(f"{output}/{id}/{id}_pred.csv")) > 0:
                continue
            p = Predict(data, f[0], model, PEFT,TRAIN_CSV)
            res = p.run()
            res.to_csv(f"{output}/{id}/{id}_pred.csv", index=False)
        total += 1
    except:
        ...
print(f"{num}/{total}")

    
```
